# 2 methods to handle missing value
- imputation/replacement
- drop

```python
# fillna()
```

```python
new_df.fillna(value = 'Missing')
```

```python
new_df.fillna(value = new_df['Age'].mean())
```

```python
new_df.fillna(method = 'ffill') # fillna is applied on entire dataset not on particular column.
```

# using loc function to impute missing values.

```python
new_df.loc[new_df['Age'].isnull()==True, 'Age']= new_df['Age'].mean()
```

```python
## when two dataframe have same columns and we need to append the rows
```
